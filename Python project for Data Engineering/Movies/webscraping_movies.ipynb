```python
# -*- coding: utf-8 -*-

import requests
import sqlite3
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://web.archive.org/web/20230902185655/https://en.everybodywiki.com/100_Most_Highly-Ranked_Films'
db_name = 'Movies.db'
table_name = 'Top_50'
csv_file = 'top_50_films.csv'
```

```python
df = pd.DataFrame(columns= ['Average_Rank','Film','Year'])
count = 0
```

```python
# get webpage
html_page = requests.get(url).text
data = BeautifulSoup(html_page, 'html.parser')
```

```python
table = data.find('tbody')
table
```

```python
rows = table.find_all('tr')
rows
```

```python
# my solution 2 (optimized)
movies_dict = {'Average_Rank':[],'Film':[],'Year':[]}

for row in rows:
    if len(row.find_all('td')) == 0:
        continue
    row
    for i,j in enumerate(movies_dict.keys()):
        movies_dict[j].append(row.find_all('td')[i].get_text())

movies_df = pd.DataFrame(movies_dict)
top_50_movies = movies_df[:50]
```

```python
top_50_movies
```

```python
# save data to csv
top_50_movies.to_csv(csv_file, index=False)
```

```python
# save data to db

conn = sqlite3.connect(db_name)
top_50_movies.to_sql(table_name,conn, if_exists= 'replace', index=False)
conn.close()
```
